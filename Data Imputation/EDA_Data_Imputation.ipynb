```python
# select numerical columns
data_numeric = df.select_dtypes(include=[np.number])
# select non-numeric columns
data_non_numeric = df.select_dtypes(include='object')
#print(non_numeric_cols)
data_numeric.shape
```

```python
df.replace(0, np.nan, inplace=True)
```

# Impute the Null values with the corresponding mean value

```python
for col in data_numeric: #consider only numerical colum and its values
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  # only do the imputation for the columns that have missing values.
        med = df[col].mean() #impute with the median
        df[col] = df[col].fillna(med)

df.shape
```
